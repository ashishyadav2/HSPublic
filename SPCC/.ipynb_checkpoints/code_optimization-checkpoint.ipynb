# Code Optimization

```python
# Enter no of expressions: 5
# Left: a
# Right: 9
# Left: b
# Right: c+d
# Left: e
# Right: c+d
# Left: f
# Right: b+e
# Left: r
# Right: :=f

expressions = []
deadcode = []
optimized = []
n = int(input("Enter no of expressions: "))
for i in range(n):
    left = input("Left: ")
    right = input("Right: ")
    expressions.append({"left": left ,"right":right})

def print_exp(array):
    for i in range(len(array)):
        print("{} = {}".format(array[i]["left"],array[i]["right"]))
        
print("\n-------------------------------------")
print("Intermediate Code")
print_exp(expressions)

print("\n-------------------------------------")
print("Dead Code Elimination")
for i in range(n-1):
    temp = expressions[i]["left"]
    for j in range(i,n):
        if temp in expressions[j]["right"]:
            deadcode.append({"left": expressions[i]["left"] ,"right":expressions[i]["right"]})

deadcode.append({"left": expressions[n-1]["left"] ,"right":expressions[n-1]["right"]})

print_exp(deadcode)
l = len(deadcode)
print("\n-------------------------------------")
print("Common Expression Elimination")
for i in range(l):
    temp = deadcode[i]["right"]
    for j in range(i+1,l):
        t = deadcode[j]["left"]
        if deadcode[j]["right"] == temp:            
            deadcode[j]["left"] = deadcode[i]["left"]
            for k in range(l):
                if t in deadcode[k]["right"]:
                    deadcode[k]["right"] = deadcode[k]["right"].replace(t,deadcode[i]["left"])
                
print_exp(deadcode)

print("\n-------------------------------------")
print("Optimized Code")
j = 1
for i in range(len(deadcode)):
    if j<len(deadcode):
        if deadcode[i]["left"] != deadcode[j]["left"] and deadcode[i]["right"] != deadcode[j]["right"]:
            optimized.append({"left": deadcode[i]["left"] ,"right":deadcode[i]["right"]})        
        j+=1

optimized.append({"left": deadcode[l-1]["left"] ,"right":deadcode[l-1]["right"]})        

print_exp(optimized)
```
